```python
#Creating new Column Request_Closing_Time using Closed_Date and Created_Date columns
dataset['Closed Date'] = pd.to_datetime(dataset['Closed Date'])
dataset['Created Date'] = pd.to_datetime(dataset['Created Date'])
dataset['Request_Closing_Time'] = dataset['Closed Date'] - dataset['Created Date']
```

```python
def to_hr(deltime):
    days = deltime.days
    hours = round(deltime.seconds/3600, 2)
    result = (days*24) + hours
    return result
```

```python
dataset['Request_Closing_hrs'] = dataset['Request_Closing_Time'].apply(to_hr)
```

```python
dataset.head()
```

```python
import math
```

```python
def service_res(hrs):
    if(math.isnan(hrs)):
        return 'Not Defined'
    elif(hrs<2.0):
        return "Excellent Service"
    elif(2.0<hrs<4.0):
        return "Good Service"
    elif(4.0<hrs<6.5):
        return "Slow Service"
    else:
        return "Bad Service"
        
```

```python
dataset['Request_Closing_Response'] = dataset['Request_Closing_hrs'].apply(service_res)
```

```python
# H0 : All Complain Types average response time, mean is similar
# H1 : Not Similar
top5_Complaint = dataset['Complaint Type'].value_counts()[:5]
top5_Complaint
```

```python
top5_Complaint_names = top5_Complaint.index
top5_Complaint_names
```

```python
test_data = dataset.loc[dataset['Complaint Type'].isin(top5_Complaint_names), ['Complaint Type', 'Request_Closing_hrs']]
test_data.head()
```

```python
test_data.isnull().sum()
```

```python
test_data.shape
```

```python
test_data.dropna(inplace = True)
```

```python
test_data.shape
```

```python
t1 = test_data[test_data['Complaint Type']== top5_Complaint_names[0]].Request_Closing_hrs
t1.head()
```

```python
t2 = test_data[test_data['Complaint Type']== top5_Complaint_names[1]].Request_Closing_hrs
t2.head()
```

```python
t3 = test_data[test_data['Complaint Type']== top5_Complaint_names[2]].Request_Closing_hrs
t3.head()
```

```python
t4 = test_data[test_data['Complaint Type']== top5_Complaint_names[3]].Request_Closing_hrs
t4.head()
```

```python
t5 = test_data[test_data['Complaint Type']== top5_Complaint_names[4]].Request_Closing_hrs
t5.head()
```

```python
import scipy.stats as stats
```

```python
stats.f_oneway(t1,t2,t3,t4,t5)
```

We can see that pvalue is less than 0.05 so we can reject null hypothesis and avg response time is not same 

```python
#H0 = Complaint Type and City is Not Related(Indepndent)
#H1 = Complaint Type and City is Related(Dependent)


top5_cities = dataset['City'].value_counts()[:5]
top5_cities
```

```python
top5_cities_names = top5_cities.index
top5_cities_names
```

```python
test_data_location = dataset.loc[(dataset['Complaint Type'].isin(top5_Complaint_names)) & (dataset['City'].isin(top5_cities_names)), ['Complaint Type','City']]
```

```python
test_data_location.head()
```

```python
pd.crosstab(test_data_location['Complaint Type'],test_data_location['City'] ,margins =True)
```

```python
ch2, p_value, df, exp_frq = stats.chi2_contingency(pd.crosstab(test_data_location['Complaint Type'], test_data_location['City']))
```

```python
print(ch2)
print(df)
print(exp_frq)
```

```python
print(p_value)
```

Hear, p-value is less than 0.05 so we can reject the null hypothesis i.e, Complaint Type and City are Related
